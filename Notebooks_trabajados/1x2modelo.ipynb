```python
import pandas as pd
from dateutil import parser

FILE_IN = "ligapro_general_limpio1_renombrado.csv"

# Cargar dataset
df = pd.read_csv(FILE_IN)

# Parsear fechas y ordenar los partidos de manera ascendente
df['fecha'] = pd.to_datetime(df['fecha'], errors='coerce', utc=False)
df = df.sort_values('fecha').reset_index(drop=True)

print("Filas:", len(df), "| Columnas:", df.shape[1])
print("Rango de fechas:", df['fecha'].min(), "→", df['fecha'].max())
df.head(10)
```

```python
# Resumen de nulos
na_total = int(df.isna().sum().sum())
print("Nulos totales:", na_total)

# Mostrar datos duplicados
dup_all = int(df.duplicated().sum())
print("Duplicados (fila completa):", dup_all)

if {'equipo_local_id','equipo_visitante_id','fecha'}.issubset(df.columns):
    dup_key = int(df.duplicated(subset=['fecha','equipo_local_id','equipo_visitante_id']).sum())
    print("Duplicados por clave (fecha, local_id, visitante_id):", dup_key)

# Tipos
print("\nTipos de datos:")
print(df.dtypes.value_counts())

# Mostrar columnas
print("\nColumnas:")
print(df.columns.tolist())
```

```python
import numpy as np
df_out = df.copy()

# Intentar obtener goles si no existen columnas específicas
def extrae_goles_desde_score(score):
    # acepta "2-1", "2 : 1", etc.
    try:
        s = str(score).replace(" ", "")
        for sep in ["-", ":", "x", "X", "/"]:
            if sep in s:
                a,b = s.split(sep)
                return int(a), int(b)
    except:
        return np.nan, np.nan
    return np.nan, np.nan

if 'goles_local' not in df_out.columns or 'goles_visitante' not in df_out.columns:
    if 'score' in df_out.columns:
        gl_gv = df_out['score'].apply(lambda s: pd.Series(extrae_goles_desde_score(s), index=['goles_local','goles_visitante']))
        df_out[['goles_local','goles_visitante']] = gl_gv
    # Si ya existían, no se tocan.

# Validar el equipo_ganador_id si hace falta
if 'equipo_ganador_id' not in df_out.columns and {'equipo_local_id','equipo_visitante_id','goles_local','goles_visitante'}.issubset(df_out.columns):
    def ganador_id(row):
        if pd.isna(row['goles_local']) or pd.isna(row['goles_visitante']):
            return np.nan
        if row['goles_local'] > row['goles_visitante']:
            return int(row['equipo_local_id'])
        elif row['goles_local'] < row['goles_visitante']:
            return int(row['equipo_visitante_id'])
        else:
            return -1  # empate
    df_out['equipo_ganador_id'] = df_out.apply(ganador_id, axis=1)

# Validar resultado_1x2 si hace falta (0=empate, 1=local, 2=visitante)
if 'resultado_1x2' not in df_out.columns and {'goles_local','goles_visitante'}.issubset(df_out.columns):
    def r1x2(row):
        if pd.isna(row['goles_local']) or pd.isna(row['goles_visitante']):
            return np.nan
        if row['goles_local'] > row['goles_visitante']: return 1
        if row['goles_local'] < row['goles_visitante']: return 2
        return 0
    df_out['resultado_1x2'] = df_out.apply(r1x2, axis=1).astype('Int64')

# Chequeo final de targets si existen
for col in ['goles_local','goles_visitante','equipo_ganador_id','resultado_1x2']:
    if col in df_out.columns:
        print(col, "→ nulos:", int(df_out[col].isna().sum()))

df_out.head(10)
```

```python
# Cargar el dataset ordenado
df = pd.read_csv("ligapro_dataset_ordenado.csv")

print(f"Shape inicial: {df.shape}")
df.head()
```

```python
# Usamos fecha + id local + id visitante como clave única
df = df.drop_duplicates(subset=["fecha", "equipo_local_id", "equipo_visitante_id"])
print(f"Shape después de eliminar duplicados: {df.shape}")
```

```python
# Identificar filas con NaN en los targets clave y filtrar las filas donde al menos uno de esos targets esté en NaN
target_cols = ["goles_local", "goles_visitante", "resultado_1x2", "equipo_ganador_id"]
nan_targets = df[df[target_cols].isna().any(axis=1)]

# Mostrar filas con NaN en los targets
if not nan_targets.empty:
    print("Partidos con NaN en los targets:")
    display(nan_targets)

# Función para parsear el score cuando está en formato "x-y"
def parse_score(score):
    try:
        g_local, g_visita = map(int, str(score).split("-"))
        return g_local, g_visita
    except:
        return None, None

# Bucle para recorrer cada fila que tenía NaN en los targets
for idx, row in nan_targets.iterrows():
    if pd.notna(row.get("score")):
        gl, gv = parse_score(row["score"])
        if gl is not None:
            df.at[idx, "goles_local"] = gl
            df.at[idx, "goles_visitante"] = gv
            # Calcular resultado_1x2 y equipo_ganador_id
            if gl > gv:
                df.at[idx, "resultado_1x2"] = 1
                df.at[idx, "equipo_ganador_id"] = row["equipo_local_id"]
            elif gl < gv:
                df.at[idx, "resultado_1x2"] = 2
                df.at[idx, "equipo_ganador_id"] = row["equipo_visitante_id"]
            else:
                df.at[idx, "resultado_1x2"] = 0
                df.at[idx, "equipo_ganador_id"] = 0  # 0 para empate

# Eliminar los datos que sigan con NaN
df = df.dropna(subset=target_cols)
print(f"Shape después de limpiar NaN en targets: {df.shape}")
```

```python
# Lista de columnas que vamos a transformar en features históricas
cols_historicas = [
    'tarjetas_amarillas_local', 'tarjetas_amarillas_visitante',
    'tarjetas_rojas_local', 'tarjetas_rojas_visitante',
    'posesion_local', 'posesion_visitante',
    'fuera_de_juego_local', 'fuera_de_juego_visitante',
    'corners_local', 'corners_visitante',
    'tiros_a_puerta_local', 'tiros_a_puerta_visitante',
    'tiros_fuera_local', 'tiros_fuera_visitante',
    'intentos_a_porteria_local', 'intentos_a_porteria_visitante',
    'atajadas_local', 'atajadas_visitante',
    'faltas_local', 'faltas_visitante',
    'penales_local', 'penales_visitante',
    'tiros_bloqueados_local', 'tiros_bloqueados_visitante',
    'ataques_peligrosos_local', 'ataques_peligrosos_visitante',
    'ataques_local', 'ataques_visitante',
    'faltas_totales', 'tarjetas_amarillas_totales', 'tarjetas_rojas_totales',
    'tarjetas_totales', 'tiros_esquina_totales'
]

# Se asegura que el DataFrame esté ordenado cronológicamente por la fecha
df = df.sort_values("fecha")

# Función que aplica shift y rolling (mirar hacia atrás y saca la media móvil con ventana definida de los 5 partidos previos)
def historico_equipo(df, col, id_col, ventana=5):
    return df.groupby(id_col)[col].shift(1).rolling(ventana, min_periods=1).mean()

# Se recorren todas las columnas históricas y se aplica la función
for col in cols_historicas:
    if col.endswith("_local"):
        # Si es de local, se agrupa por id_local
        df[col] = historico_equipo(df, col, "equipo_local_id", ventana=5)
    elif col.endswith("_visitante"):
        # Si es de visitante, se agrupa por id_visitante
        df[col] = historico_equipo(df, col, "equipo_visitante_id", ventana=5)
    else:
        # Totales: los tomamos como histórico del equipo_local_id
        df[col] = historico_equipo(df, col, "equipo_local_id", ventana=5)

# Rellenar NaN iniciales con 0
df[cols_historicas] = df[cols_historicas].fillna(0)
```

```python
import pandas as pd
import numpy as np

# Nombre del archivo con los datos históricos
FILE_IN = "ligapro_step2_historico.csv" # Dataset con los datos históricos
DATE_COL = "fecha"                      # columnas con la fecha de los partidos
TARGET_CLF = "resultado_1x2"            # target de clasificación (0=empate,1=local,2=visitante)
ID_COLS = ["equipo_local_id", "equipo_visitante_id"]  # ids de los equipos en cada partido

# Cargar el csv en un DataFrame
df = pd.read_csv(FILE_IN)

# Convertir la columna fecha a datetime
df[DATE_COL] = pd.to_datetime(df[DATE_COL], errors="coerce")

# Ordenar el DataFrame por fecha de menor a mayor y resetear índices
df = df.sort_values(DATE_COL).reset_index(drop=True)

print(df.shape, df[DATE_COL].min(), "→", df[DATE_COL].max())
```

```python
# Columnas que NO deben usarse como features
EXCLUDE_FROM_FEATURES = set([
    DATE_COL, "score", "equipo_ganador_id", "local", "visitante", "estadio",
    "goles_local", "goles_visitante",  # estos son targets de regresión; no van en el dataset de 76 columnas
    TARGET_CLF
])

# Filtro para ver las columnas numéricas (int o float) que podrían servir como features
num_cols = df.select_dtypes(include=["int64","float64"]).columns.tolist()
candidate_features = [c for c in num_cols if c not in EXCLUDE_FROM_FEATURES]

# Asegurar que los IDs de los equipos siempre estén como features
for c in ID_COLS:
    if c not in candidate_features:
        candidate_features.append(c)

# Eliminar columnas constantes
nunique = df[candidate_features].nunique()
constant_cols = nunique[nunique <= 1].index.tolist()
candidate_features = [c for c in candidate_features if c not in constant_cols]

# Total de columnas que quedaron como features
print("Candidatas:", len(candidate_features))
```

```python
TRAIN_END = "2024-12-31" # Fecha de corte para separar train y test

# Máscara booleana (todo lo que sea mennor o igual a 31/12/2024 se va a entrenamiento)
train_mask = df[DATE_COL] <= TRAIN_END
# Y lo que sea posterior al 31/12/2024 entra a test
test_mask  = df[DATE_COL] >  TRAIN_END

# Se arma el set de entrenamiento con las features candidatas y el target
X_train = df.loc[train_mask, candidate_features].copy()
y_train = df.loc[train_mask, TARGET_CLF].astype(int).copy()

# Se arma el set de test con las features candidatas y el target
X_test  = df.loc[test_mask, candidate_features].copy()
y_test  = df.loc[test_mask, TARGET_CLF].astype(int).copy()

# Relleno de NaN con 0 para evitar problemas al entrenar
X_train = X_train.fillna(0)
X_test  = X_test.fillna(0)

# Muestra las dimensiones de train y test e impresión la distribución de clases en el train
print("Train:", X_train.shape, "Test:", X_test.shape)
y_train.value_counts().sort_index()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import mutual_info_classif

# 1) Entrenamos un RandomForest para sacar importancias de features
rf = RandomForestClassifier(
    n_estimators=600,
    min_samples_leaf=2,
    class_weight="balanced",
    random_state=42,
    n_jobs=-1
)
rf.fit(X_train, y_train)
fi = pd.Series(rf.feature_importances_, index=X_train.columns)

# 2) Información mutua que captura relaciones no lineales y no monótonas
mi = pd.Series(
    mutual_info_classif(X_train, y_train, discrete_features=False, random_state=42),
    index=X_train.columns
)

# 3) Correlación de Pearson en valor absoluto (relación lineal)
pearson = pd.concat([X_train, y_train], axis=1).corr()[TARGET_CLF].drop(TARGET_CLF).abs().fillna(0)

# Normalizamos cada métrica a [0,1] basados en el ranking
def rank01(s):
    r = s.rank(method="average", ascending=True)
    return (r - r.min()) / (r.max() - r.min() + 1e-12)

# Ensemble simple de scores
score = 0.5 * rank01(fi) + 0.3 * rank01(mi) + 0.2 * rank01(pearson)
score = score.sort_values(ascending=False)

# Top de las 15 features más fuertes según el score combinado
score.head(15)
```

```python
MAX_TOTAL_COLS = 76
SLOTS_FOR_FEATURES = MAX_TOTAL_COLS - (1 + len(ID_COLS) + 1)  # fecha + IDs + target

# Forzar que los ID_COLS permanezcan
keep_ids = [c for c in ID_COLS if c in score.index]

# Quitar IDs del ranking para no duplicarlos en la selección
rank_wo_ids = score.drop(index=keep_ids, errors="ignore")

# Tomamos las top-N features según el score combinado, donde N = slots disponibles
top_feats = rank_wo_ids.index.tolist()[:SLOTS_FOR_FEATURES]

print("Slots para features:", SLOTS_FOR_FEATURES)
print("IDs que se mantienen:", keep_ids)
print("Features seleccionadas:", len(top_feats))
```

```python
FINAL_COLS = [DATE_COL] + keep_ids + top_feats + [TARGET_CLF]
df_76 = df[FINAL_COLS].copy()

# Verificamos que el número de columnas no pase el límite de 76
assert df_76.shape[1] <= MAX_TOTAL_COLS
print("Shape final:", df_76.shape)

# Guardamos el dataset final con las 76 columnas seleccionadas
OUT_76 = "ligapro_train_76.csv"
df_76.to_csv(OUT_76, index=False, encoding="utf-8")
print("Guardado:", OUT_76)
```

```python
import pandas as pd
import numpy as np
import lightgbm as lgb
from sklearn.metrics import mean_absolute_error, mean_squared_error, accuracy_score

# Configuración inicial de columnas y archivo
CSV = "ligapro_train_for_goals.csv"
DATE_COL   = "fecha"
TARGET_CLF = "resultado_1x2"
TARGET_GL  = "goles_local"
TARGET_GV  = "goles_visitante"
ID_COLS    = ["equipo_local_id", "equipo_visitante_id"]
TRAIN_END  = "2024-12-31"


# Cargar dataset y ordenar por fecha
df = pd.read_csv(CSV)
assert DATE_COL in df.columns, f"No se encontró la columna de fecha '{DATE_COL}' en {CSV}"
df[DATE_COL] = pd.to_datetime(df[DATE_COL], errors="coerce")
df = df.sort_values(DATE_COL).reset_index(drop=True)

# Validaciones básicas
for col in [TARGET_CLF, TARGET_GL, TARGET_GV] + ID_COLS:
    assert col in df.columns, f"Falta columna requerida: {col}"

# Definir features sin fuga de información
EXCLUDE = set([DATE_COL, TARGET_CLF, TARGET_GL, TARGET_GV, "score", "equipo_ganador_id", "local", "visitante", "estadio"])
FEATURES = [c for c in df.columns if c not in EXCLUDE]

# Quitar columnas constantes
nunique = df[FEATURES].nunique()
FEATURES = [c for c in FEATURES if nunique[c] > 1]

print(f"Total FEATURES: {len(FEATURES)}")

# División temporal en train y test
train_mask = df[DATE_COL] <= TRAIN_END
test_mask  = df[DATE_COL] >  TRAIN_END

print(f"Train filas: {train_mask.sum()} | Test filas: {test_mask.sum()}")

# Particiones de X (features) e Y (targets)
Xtr = df.loc[train_mask, FEATURES].fillna(0)
Xte = df.loc[test_mask,  FEATURES].fillna(0)

ytr_gl = df.loc[train_mask, TARGET_GL]
yte_gl = df.loc[test_mask,  TARGET_GL]

ytr_gv = df.loc[train_mask, TARGET_GV]
yte_gv = df.loc[test_mask,  TARGET_GV]

# Datasets de LightGBM (IDs de equipos)
categorical_feats = [c for c in ID_COLS if c in FEATURES]

dtr_gl = lgb.Dataset(Xtr, label=ytr_gl, categorical_feature=categorical_feats or None)
dva_gl = lgb.Dataset(Xte, label=yte_gl, categorical_feature=categorical_feats or None)

dtr_gv = lgb.Dataset(Xtr, label=ytr_gv, categorical_feature=categorical_feats or None)
dva_gv = lgb.Dataset(Xte, label=yte_gv, categorical_feature=categorical_feats or None)

# Parámetros para LightGBM en regresión
params_reg = dict(
    objective="regression",
    metric="mae",
    learning_rate=0.05,
    num_leaves=31,
    max_depth=6,
    min_data_in_leaf=20,
    feature_fraction=0.9,
    bagging_fraction=0.9,
    bagging_freq=1,
    random_state=42,
    verbosity=-1
)

# Callbacks para early stopping y logs
cb_early = lgb.early_stopping(stopping_rounds=100, verbose=True)
cb_log   = lgb.log_evaluation(period=200)

# Entrenamiento para goles_local
print("\n" + "="*50)
print(">> Entrenando modelo para goles_local...")
print("="*50)
model_gl = lgb.train(
    params=params_reg,
    train_set=dtr_gl,
    num_boost_round=2000,
    valid_sets=[dtr_gl, dva_gl],
    valid_names=["train","valid"],
    callbacks=[cb_early, cb_log]
)

# Entrenamiento para goles_visitante
print("\n" + "="*50)
print(">> Entrenando modelo para goles_visitante...")
print("="*50)
model_gv = lgb.train(
    params=params_reg,
    train_set=dtr_gv,
    num_boost_round=2000,
    valid_sets=[dtr_gv, dva_gv],
    valid_names=["train","valid"],
    callbacks=[cb_early, cb_log]
)

# Predicciones en Test
pred_gl = model_gl.predict(Xte, num_iteration=model_gl.best_iteration)
pred_gv = model_gv.predict(Xte, num_iteration=model_gv.best_iteration)

# Métricas para regresión
mae_gl = mean_absolute_error(yte_gl, pred_gl)
mae_gv = mean_absolute_error(yte_gv, pred_gv)
rmse_gl = np.sqrt(mean_squared_error(yte_gl, pred_gl))
rmse_gv = np.sqrt(mean_squared_error(yte_gv, pred_gv))

print("\n" + "="*50)
print("RESULTADOS EN TEST:")
print("="*50)
print(f"MAE goles_local:    {mae_gl:.4f} | RMSE: {rmse_gl:.4f}")
print(f"MAE goles_visitante: {mae_gv:.4f} | RMSE: {rmse_gv:.4f}")

# Derivar resultado 1X2 desde los goles predichos (redondeados)
gl_r = np.clip(np.round(pred_gl), 0, 15).astype(int)
gv_r = np.clip(np.round(pred_gv), 0, 15).astype(int)

def to_1x2(gl, gv):
    """Convierte marcador a resultado 1X2"""
    if gl > gv:  return 1  # Victoria local
    if gl < gv:  return 2  # Victoria visitante
    return 0               # Empate

y_pred_1x2_from_goals = np.array([to_1x2(a, b) for a, b in zip(gl_r, gv_r)])
y_true_1x2 = df.loc[test_mask, TARGET_CLF].astype(int)

# Accuracy en clasificación derivada de predicción de goles
acc_from_goals = accuracy_score(y_true_1x2, y_pred_1x2_from_goals)

# Matriz de confusión para diagnóstico
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

print("\n" + "="*50)
print(f"Accuracy 1X2 derivado del marcador (TEST): {acc_from_goals:.4f}")
print("="*50)

# Mostrar matriz de confusión
cm = confusion_matrix(y_true_1x2, y_pred_1x2_from_goals)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=["Empate", "Local", "Visitante"])
disp.plot(cmap=plt.cm.Blues)
plt.title("Matriz de Confusión (Resultado 1X2)")
plt.show()

# Guardar modelos entrenados en formato .txt
model_gl.save_model('modelo_goles_local.txt')
model_gv.save_model('modelo_goles_visitante.txt')
print("Modelos guardados: 'modelo_goles_local.txt' y 'modelo_goles_visitante.txt'")
```

```python
import pandas as pd
import numpy as np
import joblib
import optuna
from datetime import datetime
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_absolute_error, mean_squared_error, accuracy_score
from sklearn.ensemble import RandomForestClassifier
import lightgbm as lgb
from catboost import CatBoostRegressor
import warnings
warnings.filterwarnings("ignore")

CSV         = "ligapro_train_for_goals.csv"  # dataset con features + goles
DATE_COL    = "fecha"                        # columna temporal para ordenar
TARGET_CLF  = "resultado_1x2"                # target de clasificación (0 empate, 1 local, 2 visita)
TARGET_GL   = "goles_local"                  # target de regresión (goles del local)
TARGET_GV   = "goles_visitante"              # target de regresión (goles del visitante)
ID_COLS     = ["equipo_local_id", "equipo_visitante_id"]  # IDs de equipos
TRAIN_END   = "2024-12-31"                   # corte temporal anti-leak
ENSEMBLE_W  = 0.5                            # peso del ensamble
CAL_CUTS = [0.5, 1.5, 2.5, 3.5, 4.5]         # umbrales para redondeo calibrado
```

```python
# LOAD & PREP
df = pd.read_csv(CSV)
assert DATE_COL in df.columns, f"No se encontró la columna '{DATE_COL}'"
df[DATE_COL] = pd.to_datetime(df[DATE_COL], errors="coerce")
df = df.sort_values(DATE_COL).reset_index(drop=True)

# Verificar que estén las columnas clave (targets e IDs)
for col in [TARGET_CLF, TARGET_GL, TARGET_GV] + ID_COLS:
    assert col in df.columns, f"Falta columna requerida: {col}"

# Listar features evitando fuga: sacando fecha, targets y metadata textual
EXCLUDE = set([DATE_COL, TARGET_CLF, TARGET_GL, TARGET_GV,
               "score", "equipo_ganador_id", "local", "visitante", "estadio"])
FEATURES = [c for c in df.columns if c not in EXCLUDE]

# Eliminar columnas constantes
nunique = df[FEATURES].nunique()
FEATURES = [c for c in FEATURES if nunique[c] > 1]

# Split temporal
train_mask = df[DATE_COL] <= TRAIN_END
test_mask  = df[DATE_COL] >  TRAIN_END

# X con features y relleno de NaN; y separados por objetivo
Xtr = df.loc[train_mask, FEATURES].fillna(0)
Xte = df.loc[test_mask,  FEATURES].fillna(0)

ytr_gl = df.loc[train_mask, TARGET_GL].astype(float)
yte_gl = df.loc[test_mask,  TARGET_GL].astype(float)
ytr_gv = df.loc[train_mask, TARGET_GV].astype(float)
yte_gv = df.loc[test_mask,  TARGET_GV].astype(float)
ytr_1x2 = df.loc[train_mask, TARGET_CLF].astype(int)
yte_1x2 = df.loc[test_mask,  TARGET_CLF].astype(int)

# Categóricas a tratar como tales (IDs de equipos)
categorical_feats = [c for c in ID_COLS if c in FEATURES]

# Resumen del split y de las features seleccionadas
print(f"Train filas: {train_mask.sum()} | Test filas: {test_mask.sum()}")
print(f"Total FEATURES: {len(FEATURES)} | Categóricas: {categorical_feats}")
```

```python
# Validación cruzada temporal (TimeSeriesSplit) y devuelve el MAE promedio.
def mae_cv_for_params(X, y, params, n_splits=4, cats=None):
    tss = TimeSeriesSplit(n_splits=n_splits)
    maes = []
    for tr_idx, va_idx in tss.split(X):
        X_tr, X_va = X.iloc[tr_idx], X.iloc[va_idx]
        y_tr, y_va = y.iloc[tr_idx], y.iloc[va_idx]
        # Prepara los datasets para LightGBM, marcando las categóricas si hay
        dtr = lgb.Dataset(X_tr, label=y_tr, categorical_feature=cats or [])
        dva = lgb.Dataset(X_va, label=y_va, categorical_feature=cats or [])
        # Entrenamiento con early stop para no pasarse de vueltas
        model = lgb.train(
            params, dtr, num_boost_round=4000, valid_sets=[dva],
            valid_names=["valid"],
            callbacks=[lgb.early_stopping(stopping_rounds=200, verbose=False)]
        )
        # Predice en la validación y guarda el MAE
        pred = model.predict(X_va, num_iteration=model.best_iteration)
        maes.append(mean_absolute_error(y_va, pred))
    return float(np.mean(maes))

# Función de búsqueda de hiperparámetros para LightGBM (Optuna)
def make_lgb_params(trial):
    return dict(
        objective="regression",
        metric="mae",
        learning_rate=trial.suggest_float("learning_rate", 0.01, 0.1, log=True),
        num_leaves=trial.suggest_int("num_leaves", 15, 63),
        max_depth=trial.suggest_int("max_depth", 4, 10),
        min_data_in_leaf=trial.suggest_int("min_data_in_leaf", 10, 60),
        feature_fraction=trial.suggest_float("feature_fraction", 0.6, 1.0),
        bagging_fraction=trial.suggest_float("bagging_fraction", 0.6, 1.0),
        bagging_freq=1,
        lambda_l1=trial.suggest_float("lambda_l1", 0.0, 5.0),
        lambda_l2=trial.suggest_float("lambda_l2", 0.0, 5.0),
        random_state=42,
        verbosity=-1
    )

# Función de redondeo calibrado
def calibrated_round(x, cuts=CAL_CUTS):
    bins = [-np.inf] + cuts + [np.inf]
    labels = list(range(len(bins)-1))
    return int(pd.cut([x], bins=bins, labels=labels)[0])

# Función para convertir el marcador, según los goles del equipo local y visitante
def marcador_a_1x2(gl, gv):
    if gl > gv: return 1
    if gl < gv: return 2
    return 0
```

```python
# Función para goles_local
def objective_gl(trial):
    params = make_lgb_params(trial)
    return mae_cv_for_params(Xtr, ytr_gl, params, cats=categorical_feats)

print("\n[Optuna] Tuning GOLES LOCAL...")
study_gl = optuna.create_study(direction="minimize")
study_gl.optimize(objective_gl, n_trials=40)
best_params_gl = study_gl.best_params
best_params_gl.update(dict(objective="regression", metric="mae", random_state=42, verbosity=-1, bagging_freq=1))
print("Mejores params GL:", best_params_gl)

# Función para goles_vistante
def objective_gv(trial):
    params = make_lgb_params(trial)
    return mae_cv_for_params(Xtr, ytr_gv, params, cats=categorical_feats)

print("\n[Optuna] Tuning GOLES VISITANTE...")
study_gv = optuna.create_study(direction="minimize")
study_gv.optimize(objective_gv, n_trials=40)
best_params_gv = study_gv.best_params
best_params_gv.update(dict(objective="regression", metric="mae", random_state=42, verbosity=-1, bagging_freq=1))
print("Mejores params GV:", best_params_gv)
```

```python
# Entrenamientos finales
dtr_gl = lgb.Dataset(Xtr, label=ytr_gl, categorical_feature=categorical_feats or [])
dtr_gv = lgb.Dataset(Xtr, label=ytr_gv, categorical_feature=categorical_feats or [])

# Entrenamiento final de LightGBM (goles_local) con early stopping para no sobreajustar
print("\nEntrenando LGBM final GL...")
model_gl_lgb = lgb.train(best_params_gl, dtr_gl, num_boost_round=6000,
                         valid_sets=[dtr_gl], valid_names=["train"],
                         callbacks=[lgb.early_stopping(stopping_rounds=400, verbose=False)])

# Entrenamiento final de LightGBM (goles_visitante) con early stopping para no sobreajustar
print("Entrenando LGBM final GV...")
model_gv_lgb = lgb.train(best_params_gv, dtr_gv, num_boost_round=6000,
                         valid_sets=[dtr_gv], valid_names=["train"],
                         callbacks=[lgb.early_stopping(stopping_rounds=400, verbose=False)])

# CatBoost (segundo modelo para el ensamble)
print("\nEntrenando CatBoost...")
cat_params = dict(loss_function="MAE", depth=6, learning_rate=0.05, l2_leaf_reg=3.0,
                  iterations=3000, random_seed=42, verbose=False)
model_gl_cat = CatBoostRegressor(**cat_params)
model_gv_cat = CatBoostRegressor(**cat_params)
model_gl_cat.fit(Xtr, ytr_gl, cat_features=[Xtr.columns.get_loc(c) for c in categorical_feats] if categorical_feats else None)
model_gv_cat.fit(Xtr, ytr_gv, cat_features=[Xtr.columns.get_loc(c) for c in categorical_feats] if categorical_feats else None)

# Ensemble simple con promedio ponderado entre LGBM y CatBoost
pred_gl = ENSEMBLE_W * model_gl_lgb.predict(Xte) + (1-ENSEMBLE_W) * model_gl_cat.predict(Xte)
pred_gv = ENSEMBLE_W * model_gv_lgb.predict(Xte) + (1-ENSEMBLE_W) * model_gv_cat.predict(Xte)

# Métricas de regresión en test
mae_gl = mean_absolute_error(yte_gl, pred_gl)
mae_gv = mean_absolute_error(yte_gv, pred_gv)
rmse_gl = np.sqrt(mean_squared_error(yte_gl, pred_gl))
rmse_gv = np.sqrt(mean_squared_error(yte_gv, pred_gv))

print("\n================ RESULTADOS EN TEST ================")
print(f"MAE goles_local: {mae_gl:.4f} | RMSE: {rmse_gl:.4f}")
print(f"MAE goles_visitante: {mae_gv:.4f} | RMSE: {rmse_gv:.4f}")

# Redondeo calibrado de los goles para luego derivar 1X2
gl_r = np.array([calibrated_round(x, CAL_CUTS) for x in pred_gl])
gv_r = np.array([calibrated_round(x, CAL_CUTS) for x in pred_gv])

# Accuracy de 1X2 obtenido a partir del marcador predicho
acc_from_goals = accuracy_score(yte_1x2, [marcador_a_1x2(a,b) for a,b in zip(gl_r, gv_r)])
print(f"Accuracy 1X2 derivado del marcador: {acc_from_goals:.4f}")
```

```python
# ==================== STACKING =====================
print("\nGenerando OOF para stacking 1X2...")
tss = TimeSeriesSplit(n_splits=4) # CV temporal con 4 folds
oof_gl = np.zeros(len(Xtr))       # Predcciones de goles local
oof_gv = np.zeros(len(Xtr))       # Predcciones de goles visitante
for tr_idx, va_idx in tss.split(Xtr):
    X_tr, X_va = Xtr.iloc[tr_idx], Xtr.iloc[va_idx] # split temporal (train y valid por fold)
    y_tr_gl_, y_tr_gv_ = ytr_gl.iloc[tr_idx], ytr_gv.iloc[tr_idx] # targets para ese fold

    # Entrenamiento de modelos LGBM con los mejores params por fold y obtener predicciones en la parte de validación
    _gl = lgb.train(best_params_gl, lgb.Dataset(X_tr, label=y_tr_gl_), num_boost_round=2000)
    _gv = lgb.train(best_params_gv, lgb.Dataset(X_tr, label=y_tr_gv_), num_boost_round=2000)
    # Guardar las predicciones en los índices de validación
    oof_gl[va_idx] = _gl.predict(X_va)
    oof_gv[va_idx] = _gv.predict(X_va)

# armo datasets para el meta-clasificador para train y preds de test para inferencia
stack_train = pd.DataFrame({"pred_gl": oof_gl, "pred_gv": oof_gv}, index=Xtr.index)
stack_test  = pd.DataFrame({"pred_gl": pred_gl, "pred_gv": pred_gv}, index=Xte.index)

# Modelo de stacking RandomForest que aprende a mapear (pred_gl, pred_gv) -> 1X2
clf = RandomForestClassifier(n_estimators=600, max_depth=8, min_samples_leaf=2,
                             class_weight="balanced", random_state=42, n_jobs=-1)
clf.fit(stack_train, ytr_1x2)
acc_stack = accuracy_score(yte_1x2, clf.predict(stack_test))

# elijo el modo final si el stacking gana o empata, me quedo con stacking, si no con el 1X2 derivado de goles
FINAL_MODE = "stacking" if acc_stack >= acc_from_goals else "from_goals"

print("\n================= RESUMEN FINAL =================")
print(f"MAE goles_local:     {mae_gl:.4f} | RMSE: {rmse_gl:.4f}")
print(f"MAE goles_visitante: {mae_gv:.4f} | RMSE: {rmse_gv:.4f}")
print(f"Accuracy 1X2 derivado del marcador: {acc_from_goals:.4f}")
print(f"Accuracy 1X2 con stacking:          {acc_stack:.4f}")
print(f"Modo recomendado para predicción final: {FINAL_MODE}")
print("==================================================")
```
